## Funciones

```python
#Función para obtener los canales RGB de una imagen
def RGB(img):
    img = img.astype(float) #Cambio tipo de dato para trabajar en float

    #Obtengo las dimensiones de la imagen 
    dimension = img.shape

    #Creo matrices vacias
    r = np.zeros((dimension[0],dimension[1]), dtype = float)
    g = np.zeros((dimension[0],dimension[1]), dtype = float)
    b = np.zeros((dimension[0],dimension[1]), dtype = float)

    #Obtengo componentes RGB
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]

    #Clipeo los componentes RGB
    r = np.clip(r,0,255)
    g = np.clip(g,0,255)
    b = np.clip(b,0,255)

    return r,g,b

#Función para clipear la luminancia
def clipY(mat):
    mat = np.clip(mat,0.0,1.0)
    return mat

#Función para clipear la componente en fase
def clipI(mat):
    mat = np.clip(mat,-0.5957,0.5957)          
    return mat

#Función para clipear la el componente en cuadratura
def clipQ(mat):
    mat = np.clip(mat,-0.5226,0.5226)          
    return mat

#Función para obtener YIQ a partir de RGB
def RGB2YIQ(img):
    img = img.astype(float) #Cambio tipo de dato para trabajar en float

    #obtengo los componentes RGB
    r,g,b = RGB(img)

    #Normalizacion
    r = r/255 
    g = g/255 
    b = b/255 

    #Obtengo las dimensniones de la imagen
    dimension = img.shape

    #Creo matrices con zeros 
    Y = np.zeros((dimension[0],dimension[1]), dtype = float)
    I = np.zeros((dimension[0],dimension[1]), dtype = float)
    Q = np.zeros((dimension[0],dimension[1]), dtype = float)

    #Obtengo matrices YIQ
    Y = (r*0.299+g*0.587+b*0.114)
    I = (r*0.595716-g*0.274453-b*0.321263)
    Q = (r*0.211456-g*0.522591+b*0.311135)

    #Clipeo las matrices YIQ a sus respectivos rangos
    Y = clipY(Y)
    I = clipI(I)
    Q = clipQ(Q)

    return(Y,I,Q)



#Función para obtener una imagen RGB a partir de YIQ 
def YIQ2RGB(Y,I,Q):
    Y = clipY(Y) #función para clipear luminancia
    I = clipI(I) #función para clipear luminancia
    Q = clipQ(Q) #función para clipear luminancia

    #Obtengo las dimensiones de la imagen
    dimension = Y.shape

    #Creo matrices para componentes RGB vacias
    r = np.zeros((dimension[0],dimension[1]), dtype = float)
    g = np.zeros((dimension[0],dimension[1]), dtype = float)
    b = np.zeros((dimension[0],dimension[1]), dtype = float)

    #Asigno los valores a los canales RGB
    r = (Y+0.9563*I+0.621*Q)*255
    g = (Y-0.2721*I-0.6474*Q)*255
    b = (Y-1.1070*I+1.7046*Q)*255

    #Clipeo los componentes RGB
    r = np.clip(r,0,255)
    g = np.clip(g,0,255)
    b = np.clip(b,0,255)
    
    img = np.zeros((dimension[0],dimension[1],3),dtype = 'uint8') #Transformo a tipo de dato uint8 ya que es el requerido para mostrar las imágenes
    img[:,:,0] = r
    img[:,:,1] = g
    img[:,:,2] = b

    return(img)


############################ FUNCIONES PARA CARGAR IMAGEN ######################

def cargarImg1(self,image):
    frame = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    imagen = QImage(frame, frame.shape[1], frame.shape[0], frame.strides[0], QImage.Format_RGB888)
    imagen = imagen.scaled(400, 280, Qt.KeepAspectRatio)
    self.imagen1.setPixmap(QtGui.QPixmap.fromImage(imagen))

def cargarImg2(self,image):
    frame = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    imagen = QImage(frame, frame.shape[1], frame.shape[0], frame.strides[0], QImage.Format_RGB888)
    imagen = imagen.scaled(400, 280, Qt.KeepAspectRatio)
    self.imagen2.setPixmap(QtGui.QPixmap.fromImage(imagen))

def cargarImg3(self, image):
    frame = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    imagen = QImage(frame, frame.shape[1], frame.shape[0], frame.strides[0], QImage.Format_RGB888)
    imagen = imagen.scaled(400, 280, Qt.KeepAspectRatio)
    self.imagen3.setPixmap(QtGui.QPixmap.fromImage(imagen))

############################ Elegir el tipo de operacion  ######################
    
def operaciones(operation, imgAmod, imgBmod):
        
    if operation == "Suma Clampeada RGB":
        return sumaRGBClampeada(imgAmod, imgBmod)

    elif operation == "Suma Promediada RGB":
        return sumaRGBProm(imgAmod, imgBmod)

    elif operation == "Suma Clampeada YIQ":
        return sumaYIQClampeada(imgAmod, imgBmod)

    elif operation == "Suma Promediada YIQ":
        return sumaYIQProm(imgAmod, imgBmod)

    elif operation == "Resta Clampeada RGB":
        return restaRGBClampeada(imgAmod, imgBmod)

    elif operation == "Resta Promediada RGB":
        return restaRGBProm(imgAmod, imgBmod)

    elif operation == "Resta Clampeada YIQ":
        return restaRGBClampeada(imgAmod, imgBmod)

    elif operation == "Resta Promediada YIQ":
        return restaYIQProm(imgAmod, imgBmod)
    
    elif operation == "Producto":
        return producto(imgAmod, imgBmod)
    
    elif operation == "Resta Valor Absoluto":
        return restaABS(imgAmod, imgBmod)
    
    elif operation == "Cociente":
        return cociente(imgAmod, imgBmod)
    
    elif operation == "If Darker":
        return YIQifDarker(imgAmod, imgBmod)
    
    elif operation == "If Lighter":
        return YIQifLighter(imgAmod, imgBmod)
  
```

### 1- Cuasi suma y resta (clampeada y promediada) en el espacio de color RGB

```python

#Función para realizar la operación suma RGB Clampeada
def sumaRGBClampeada(imgAmod,imgBmod):
    RA,GA,BA = RGB(imgAmod)
    RB,GB,BB = RGB(imgBmod)
   
    RC = RA+RB
    GC = GA+GB
    BC = BA+BB

    RC = np.clip(RC,0,255)
    GC = np.clip(GC,0,255)
    BC = np.clip(BC,0,255)

    img = np.zeros((imgAmod.shape[0],imgAmod.shape[1],3),dtype = 'uint8')
    img[:,:,0] = RC
    img[:,:,1] = GC
    img[:,:,2] = BC

    return(img)

#Función para realizar la operación resta RGB Clampeada
def restaRGBClampeada(imgAmod,imgBmod):
    RA,GA,BA = RGB(imgAmod)
    RB,GB,BB = RGB(imgBmod)

    RC = RA-RB
    GC = GA-GB
    BC = BA-BB

    RC = np.clip(RC,0,255)
    GC = np.clip(GC,0,255)
    BC = np.clip(BC,0,255)
    
    dimAmod = imgAmod.shape
    dimBmod = imgBmod.shape
    
    img = np.zeros((imgAmod.shape[0],imgAmod.shape[1],3),dtype = 'uint8')
    img[:,:,0] = RC
    img[:,:,1] = GC
    img[:,:,2] = BC

    return(img)
           
#Función para realizar la operación suma RGB promediada
def sumaRGBProm(imgAmod, imgBmod):
    RA,GA,BA = RGB(imgAmod)
    RB,GB,BB = RGB(imgBmod)

    RC = (RA+RB)//2
    GC = (GA+GB)//2
    BC = (BA+BB)//2

    img = np.zeros((imgAmod.shape[0],imgAmod.shape[1],3),dtype = 'uint8')
    img[:,:,0] = RC
    img[:,:,1] = GC
    img[:,:,2] = BC

    return(img)

#Función para realizar la operación resta RGB promediada
def restaRGBProm(imgAmod, imgBmod):
    RA,GA,BA = RGB(imgAmod)
    RB,GB,BB = RGB(imgBmod)

    RC = (RA-RB)//2
    GC = (GA-GB)//2
    BC = (BA-BB)//2

    img = np.zeros((imgAmod.shape[0],imgAmod.shape[1],3),dtype = 'uint8')
    img[:,:,0] = RC
    img[:,:,1] = GC
    img[:,:,2] = BC

    return(img)
```

### 2- Ídem punto 1 en el espacio de color YIQ

```python
#Función para realizar la operación suma YIQ Clampeada
def sumaYIQClampeada(imgAmod, imgBmod):
    YA,IA,QA = RGB2YIQ(imgAmod)
    YB,IB,QB = RGB2YIQ(imgBmod)

    YC = YA+YB

    YC = np.clip(YC,0,1)
    
    IC = (YA*IA+YB*IB)/(YA+YB)
    QC = (YA*QA+YB*QB)/(YA+YB)
    
    imgC = YIQ2RGB(YC,IC,QC)
    return(imgC)

#Función para realizar la operación resta YIQ Clampeada
def restaYIQClampeada(imgAmod, imgBmod):
    YA,IA,QA = RGB2YIQ(imgAmod)
    YB,IB,QB = RGB2YIQ(imgBmod)

    YC = YA-YB

    YC = np.clip(YC,0,1)
    
    IC = (YA*IA-YB*IB)/(YA+YB)
    QC = (YA*QA-YB*QB)/(YA+YB)
    
    imgC = YIQ2RGB(YC,IC,QC)
    return(imgC)

#Función para realizar la operación suma YIQ promediada
def sumaYIQProm(imgAmod, imgBmod):
    YA,IA,QA = RGB2YIQ(imgAmod)
    YB,IB,QB = RGB2YIQ(imgBmod)

    YC = (YA+YB)/2

    IC = (YA*IA+YB*IB)/(YA+YB)
    QC = (YA*QA+YB*QB)/(YA+YB)
    
    imgC = YIQ2RGB(YC,IC,QC)
    return(imgC)

#Función para realizar la operación resta YIQ promediada
def restaYIQProm(imgAmod, imgBmod):
    YA,IA,QA = RGB2YIQ(imgAmod)
    YB,IB,QB = RGB2YIQ(imgBmod)

    YC = (YA-YB)/2

    IC = (YA*IA-YB*IB)/(YA+YB)
    QC = (YA*QA-YB*QB)/(YA+YB)
    
    imgC = YIQ2RGB(YC,IC,QC)
    return(imgC)
```

### 3- Producto y cociente de dos imágenes

```python
def producto(imgAmod, imgBmod):
    
    RA,GA,BA = RGB(imgAmod)
    RB,GB,BB = RGB(imgBmod)

    RC = np.multiply(RA,RB)
    GC = np.multiply(GA,GB)
    BC = np.multiply(BA,BB)

    RC = np.clip(RC,0,255)
    GC = np.clip(GC,0,255)
    BC = np.clip(BC,0,255)
    
    dimAmod = imgAmod.shape
    dimBmod = imgBmod.shape
    
    img = np.zeros((imgAmod.shape[0],imgAmod.shape[1],3),dtype = 'uint8')
    img[:,:,0] = RC
    img[:,:,1] = GC
    img[:,:,2] = BC
    
    return (img)

def cociente(imgAmod, imgBmod):
    
    RA,GA,BA = RGB(imgAmod)
    RB,GB,BB = RGB(imgBmod)
    
    #En caso de div por 0 se Aproxima con 0.0001
    RB = np.clip(RB,0.0001,255)
    GB = np.clip(RB,0.0001,255)
    BB = np.clip(RB,0.0001,255)
    
    RC = np.divide(RA, RB)
    GC = np.divide(GA, GB)
    BC = np.divide(BA, BB)

    RC = np.clip(RC,0,255)
    GC = np.clip(GC,0,255)
    BC = np.clip(BC,0,255)
    
    dimAmod = imgAmod.shape
    dimBmod = imgBmod.shape
    
    img = np.zeros((imgAmod.shape[0],imgAmod.shape[1],3),dtype = 'uint8')
    img[:,:,0] = RC
    img[:,:,1] = GC
    img[:,:,2] = BC
    
    return (img)
```

### 4- Resta de dos imágenes implementando el valor absoluto

```python
def restaABS(imgAmod,imgBmod):
    RA,GA,BA = RGB(imgAmod)
    RB,GB,BB = RGB(imgBmod)

    RC = abs(RA-RB)
    GC = abs(GA-GB)
    BC = abs(BA-BB)

    RC = np.clip(RC,0,255)
    GC = np.clip(GC,0,255)
    BC = np.clip(BC,0,255)
    
    dimAmod = imgAmod.shape
    dimBmod = imgBmod.shape
    
    img = np.zeros((imgAmod.shape[0],imgAmod.shape[1],3),dtype = 'uint8')
    img[:,:,0] = RC
    img[:,:,1] = GC
    img[:,:,2] = BC

    return(img)
```

### 5- ‘If- darker ’e ‘ If ligther’

```python
#Función para realizar la operación if darker en YIQ
def YIQifDarker(imgAmod, imgBmod):
    YA,IA,QA = RGB2YIQ(imgAmod)
    YB,IB,QB = RGB2YIQ(imgBmod)
    YC = np.zeros([imgAmod.shape[0],imgAmod.shape[1]])
    IC = np.zeros([imgAmod.shape[0],imgAmod.shape[1]])
    QC = np.zeros([imgAmod.shape[0],imgAmod.shape[1]])

    for i in range(imgAmod.shape[0]):
        for j in range(imgAmod.shape[1]):
            if YA[i,j]>=YB[i,j]:
                YC[i,j] = YB[i,j]
                IC[i,j] = IB[i,j]
                QC[i,j] = QB[i,j]
            else:
                YC[i,j] = YA[i,j]
                IC[i,j] = IA[i,j]
                QC[i,j] = QA[i,j]

    imgC = YIQ2RGB(YC,IC,QC)
    return(imgC)

#Función para realizar la operación if lighter en YIQ
def YIQifLighter(imgAmod, imgBmod):
    YA,IA,QA = RGB2YIQ(imgAmod)
    YB,IB,QB = RGB2YIQ(imgBmod)

    YC = np.zeros([imgAmod.shape[0],imgAmod.shape[1]])
    IC = np.zeros([imgAmod.shape[0],imgAmod.shape[1]])
    QC = np.zeros([imgAmod.shape[0],imgAmod.shape[1]])

    for i in range(imgAmod.shape[0]):
        for j in range(imgAmod.shape[1]):
            if YA[i,j]>=YB[i,j]:
                YC[i,j] = YA[i,j]
                IC[i,j] = IA[i,j]
                QC[i,j] = QA[i,j]
            else:
                YC[i,j] = YB[i,j]
                IC[i,j] = IB[i,j]
                QC[i,j] = QB[i,j]

    imgC = YIQ2RGB(YC,IC,QC)
    return(imgC)

```
